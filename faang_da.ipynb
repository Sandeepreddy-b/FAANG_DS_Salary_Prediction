```python
df4 = df.copy()
```

```python
loc_max_sal = df4.groupby('location')['totalyearlycompensation'].agg(['max'])
loc_min_sal = df4.groupby('location')['totalyearlycompensation'].agg(['min'])

loc_max_sal = loc_max_sal.reset_index()
loc_min_sal = loc_min_sal.reset_index()

faang_loc_sal = pd.concat([loc_max_sal, loc_min_sal], axis=1)

faang_loc_sal = faang_loc_sal.loc[:,~faang_loc_sal.columns.duplicated()]

faang_loc_sal
```

```python
loc_list = faang_loc_sal['location'].tolist()
loc_max = faang_loc_sal['max'].tolist()
loc_min = faang_loc_sal['min'].tolist()
```

```python
app = dash.Dash(__name__)
```

```python
def make_break(num_breaks):
    br_list = [html.Br()] * num_breaks
    return br_list
            

app.layout = html.Div([
    
    html.H1("FAANG Data Science Job's Analysis",style = {'color':'white', 'text-align':'center',
                                                'background-color': 'CornflowerBlue',
                                                'border-style': 'double','font-family':'Monaco'}),
    
    *make_break(2),
    
    html.Img(src='data:image/png;base64,{}'.format(encoded_faang.decode()),style={'width':'100%'}),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': ye, 'y': jobs, 'type': 'bar'}],
            'layout': {'title': 'Jobs Per Year'}
                }
            ),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': c, 'y': j, 'type': 'bar'}],
            'layout': {'title': 'Jobs By Company'}
                }
            ),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': lo, 'y': j2, 'type': 'bar'}],
            'layout': {'title': 'Jobs BY Location','auto':'True','height':'10000px'}
                }
            ),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': c, 'y': max_, 'type': 'bar','name':'Maximum Salary'},
                    {'x': c, 'y': min_, 'type': 'bar','name':'Minimum Salary'}],
            'layout': {'title': 'Salaries By Company'}
                }
            ),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': loc_list, 'y': loc_max, 'type': 'bar','name':'Maximum Salary'},
                    {'x': loc_list, 'y': loc_min, 'type': 'bar','name':'Minimum Salary'}],
            'layout': {'title': 'Salaries By Locations'}
                }
            ),
    
    *make_break(2),
    
    dcc.Graph(
        figure={
            'data': [{'x': exp_years, 'y': exp_max, 'type': 'bar','name':'Maximum Salary'},
                    {'x': exp_years, 'y': exp_min, 'type': 'bar','name':'Minimum Salary'}],
            'layout': {'title': 'Salaries By No.of Years of Experience'}
                }
            ),
    
    
    *make_break(5),
    
    html.Div('Source:https://www.kaggle.com/jackogozaly/data-science-and-stem-salaries',style={'color':'white', 
                                                                                               'text-align':'center',
                                                                                               'background-color': 'CornflowerBlue',
                                                                                               'border-style': 'double',
                                                                                               'font-family':'Monaco'})
])
```

```python
if __name__ == '__main__':
    app.run_server()
```
